```python
df=pd.read_excel("microretailer_mit_lift_lab.xlsx")
df.head(5)
```

```python
df_2= ['_updated_by','232_type_of_store','99_does_the_micro_retailer_exhibits_products_outside_','102_does_the_micro_retailer_sells_fresh_products_','184_store_devices','5_change_store_space_last_year','6_change_employees_average_salary_last_year','18_sales_records','145_number_direct_competitors','186_internet_connection', '49_inventory_records','210_sales_channels','189_payment_methods','214_customer_relationship_tools','193_sales_planning_tools','60_services','69_why_not_services','158_charact_organize_products_on_shelves','161_actions_stockouts','234_willingness_participate_workshop']
df2= df[df_2]
df2.head(5)
```

```python
total_nulos=df2.isnull().sum()
total_nulos
```

```python
df_retail=df2.fillna('Unknown')
```

```python
df_retail.info()
```

```python
tabla1 = df_retail['5_change_store_space_last_year'].value_counts()
# Crear el grfico de barras
plt.figure(figsize=(10, 6))
tabla1.plot(kind='bar', rot=45, color='skyblue')
plt.xlabel('Cambios en el Espacio de la Tienda el ltimo Ao')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Cambios en el Espacio de la Tienda el ltimo Ao')
plt.show()
```

```python
tabla2= freq_tbl(df_retail['232_type_of_store'])
#Realizamos grafico de rea pero solo filtramos los que tengan una frecuencia mayor a 1
filtro2= tabla2[tabla2['frequency']>1]
Filtro2_index= filtro2.set_index('232_type_of_store')

ax=Filtro2_index.plot(kind='area', figsize=(10,4),alpha = 0.5)
plt.show()
```

```python
tabla3 = df_retail['186_internet_connection'].value_counts(normalize=True) * 100
plt.figure(figsize=(8, 8))
plt.pie(tabla3, labels=tabla3.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Porcentaje de Conexin a Internet')
plt.show()
```

```python
tabla4 = df_retail['102_does_the_micro_retailer_sells_fresh_products_'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(tabla4, labels=tabla4.index, autopct='%1.1f%%', startangle=140)
plt.title('Porcentaje de Venta de Productos Frescos en Micro Minoristas')
plt.show()
```

```python
frecuencias_productos = df_retail['184_store_devices'].value_counts()
tabla5 = frecuencias_productos.head(5)
plt.figure(figsize=(10, 6))
plt.bar(tabla5.index, tabla5.values, color='red')
plt.xlabel('Dispositivos')
plt.ylabel('Frecuencia')
plt.title('Los 5 dispositivos ms utilizados')
plt.xticks(rotation=45)
plt.show()
```

```python
tabla6=freq_tbl(df_retail['_updated_by'])
tabla6
Filtro1= tabla6.set_index('_updated_by')
top_20_categories = Filtro1.head(15)
top_20_categories['frequency'].plot(kind='pie', figsize=(10,5), shadow=True, autopct="%0.1f %%")
plt.title('Grfico de las personas con mayores actualizaciones')
plt.show()
```
